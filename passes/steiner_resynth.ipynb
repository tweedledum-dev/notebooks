```python
from tweedledum.target import Device
from tweedledum.synthesis import a_star_swap_synth

path = Device.path(8)
circuit = a_star_swap_synth(path, [0, 1, 2, 3, 4, 5, 6, 7], [7, 6, 5, 4, 3, 2, 1, 0])
print(circuit)
print("CX cost:", 3 * len(circuit))
```

```python
from tweedledum.passes import steiner_resynth

opt = steiner_resynth(path, circuit)
print(opt)
print("CX cost:", len(opt))
```
